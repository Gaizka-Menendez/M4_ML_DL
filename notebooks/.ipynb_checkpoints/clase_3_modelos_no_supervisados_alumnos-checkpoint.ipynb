Chuleta (cheatsheet) Algoritmos:

| Algoritmo                      | Tipo                               | Hiperpar치metros principales                | Requiere preprocesado?                          | 쮺u치ndo usarlo?                                                            | Ventajas                                                         | Inconvenientes                                                                            |
| ------------------------------ | ---------------------------------- | ------------------------------------------ | ------------------------------------------------ | -------------------------------------------------------------------------- | ---------------------------------------------------------------- | ----------------------------------------------------------------------------------------- |
| **K-Means**                    | Particional                        | - n\_clusters (k)<br>- init<br>- max\_iter | Escalado de variables recomendado                | Cuando esperas grupos de forma "redondeada" y tienes idea del n췈 de grupos | F치cil de entender y r치pido                                       | Sensible a outliers y al escalado; no detecta bien formas no esf칠ricas; hay que indicar `k` |
| **DBSCAN**                     | Basado en densidad                 | - eps (풧)<br>- min\_samples (d)            | Escalado recomendable<br>(depende de la m칠trica) | Cuando no sabes el n췈 de grupos o hay ruido; grupos de cualquier forma     | Detecta ruido/outliers, no requiere `k`, grupos de cualquier forma | Puede fallar con densidades variables; definir `eps` es complicado                          |
| **HDBSCAN**                    | Basado en densidad jer치rquica        | - min\_cluster\_size<br>- min\_samples     | Escalado recomendable   | Clusters de distinta densidad, encontrar ruido, sin saber `k`   | No requiere `k`, detecta outliers, mejor con densidades distintas que DBSCAN | Lento en grandes datasets, menos interpretabilidad, m치s complejo de configurar   |
| **Agglomerative (Jer치rquico)** | Jer치rquico                         | - linkage<br>- n\_clusters                 | Escalado recomendable                            | Cuando quieres visualizar la estructura jer치rquica o no sabes k            | No requiere definir `k` a priori, visualizaci칩n con dendrograma    | Costoso en memoria con muchos datos, sensible a la m칠trica y outliers                     |
| **Mean Shift**                 | Basado en densidad                 | - bandwidth                                | Escalado recomendable                            | Cuando no sabes el n췈 de clusters y buscas modos densos en los datos       | Detecta clusters de cualquier forma y cantidad                   | Lento en grandes datasets; elecci칩n de `bandwidth` dif칤cil                                  |
| **Gaussian Mixture (GMM)**     | Modelos probabil칤sticos            | - n\_components<br>- covariance\_type      | Escalado muy recomendable                        | Cuando los grupos pueden solaparse o tienen forma el칤ptica                 | Permite solapamiento y probabilidades suaves de pertenencia      | M치s lento, sensible a inicializaci칩n y outliers; hay que indicar n췈 de componentes        |
| **Spectral Clustering**        | Basado en grafos                   | - n\_clusters<br>- affinity                | Escalado recomendado                             | Cuando la forma de los clusters es compleja/no lineal                      | Puede encontrar clusters de formas complicadas                   | No escala bien a grandes datasets; requiere saber `k`; sensible a la afinidad elegida       |


### Reducci칩n de la dimensionalidad

| Algoritmo                                     | Tipo                      | Hiperpar치metros principales                                   | Requiere preprocesado?                                | 쮺u치ndo usarlo?                                                                               | Ventajas                                                                 | Inconvenientes                                                                          |
| --------------------------------------------- | ------------------------- | ------------------------------------------------------------- | ------------------------------------------------------ | --------------------------------------------------------------------------------------------- | ------------------------------------------------------------------------ | --------------------------------------------------------------------------------------- |
| **PCA (An치lisis de Componentes Principales)** | Lineal                    | - n\_components                                               | Escalado **muy recomendable**<br>(centrado en media 0) | Cuando los datos tienen correlaci칩n lineal, para visualizaci칩n o preprocesado                 | R치pido, f치cil de interpretar, 칰til para compresi칩n y visualizaci칩n 2D/3D | Solo capta relaciones lineales; puede perder interpretabilidad                          |
| **t-SNE**                                     | No lineal (proximidad)    | - n\_components<br>- perplexity<br>- learning\_rate           | Escalado recomendable                                  | Para visualizar datos complejos en 2D o 3D, clusters no lineales (im치genes, textos, gen칠tica) | Descubre estructuras complejas; visualizaci칩n clara de grupos            | Muy lento para muchos datos (>10.000); resultados pueden variar; dif칤cil de interpretar |
| **UMAP**                                      | No lineal (proximidad)    | - n\_neighbors<br>- min\_dist<br>- n\_components              | Escalado recomendable                                  | Alternativa r치pida a t-SNE, para visualizaci칩n o reducci칩n previa al clustering               | Muy r치pido, preserva bien la estructura de los datos                     | Los resultados pueden variar seg칰n los hiperpar치metros; menos interpretable             |
| **Autoencoder**                               | No lineal (deep learning) | - arquitectura (capas, neuronas)<br>- epochs<br>- latent\_dim | Normalizaci칩n o estandarizaci칩n                        | Cuando hay muchos datos, relaciones muy complejas y se usan redes neuronales                  | Puede captar relaciones muy complejas; flexible                          | Requiere m치s datos, entrenamiento y conocimientos de deep learning                      |

### Definici칩n de funciones

````python
def comparar_clustering_escalado(
    df: pd.DataFrame, 
    columna_x: str, 
    columna_y: str, 
    columna_clase: str, 
    n_clusters: int = 3, 
    random_state: int = 42
) -> go.Figure:
    """
    Crea un subplot para comparar clustering K-means con diferentes m칠todos de escalado.
    
    Args:
        df (pd.DataFrame): DataFrame con los datos
        columna_x (str): Nombre de la columna para el eje X
        columna_y (str): Nombre de la columna para el eje Y
        columna_clase (str): Nombre de la columna con las clases reales
        n_clusters (int): N칰mero de clusters para K-means
        random_state (int): Semilla aleatoria para reproducibilidad
        
    Returns:
        go.Figure: Figura con los 4 subplots
    """
    
    # Crear figura con subplots 2x2
    fig = make_subplots(
        rows=2, cols=2,
        subplot_titles=(
            "Clases Reales",
            "Clusters K-means (Sin escalar)",
            "Clusters K-means (StandardScaler)",
            "Clusters K-means (MinMaxScaler)"
        )
    )
    
    # Preparar datos para clustering (solo las dos columnas seleccionadas)
    X = df[[columna_x, columna_y]].values
    
    # Crear paleta de colores
    colores = ['#e41a1c', '#377eb8', '#4daf4a', '#984ea3', '#ff7f00']
    
    # 1. Gr치fico de clases reales (arriba a la izquierda)
    clases_unicas = df[columna_clase].unique()
    for i, clase in enumerate(clases_unicas):
        mask = df[columna_clase] == clase
        fig.add_trace(
            go.Scatter(
                x=df.loc[mask, columna_x],
                y=df.loc[mask, columna_y],
                mode='markers',
                marker=dict(color=colores[i % len(colores)], size=8),
                name=f'Clase: {clase}',
                legendgroup='clases',
                showlegend=True
            ),
            row=1, col=1
        )
    
    # 2. K-means sin escalar (arriba a la derecha)
    kmeans_no_scaling = KMeans(n_clusters=n_clusters, random_state=random_state)
    clusters_no_scaling = kmeans_no_scaling.fit_predict(X)
    centroides_no_scaling = kmeans_no_scaling.cluster_centers_
    
    for i in range(n_clusters):
        mask = clusters_no_scaling == i
        fig.add_trace(
            go.Scatter(
                x=X[mask, 0],
                y=X[mask, 1],
                mode='markers',
                marker=dict(color=colores[i % len(colores)], size=8),
                name=f'Cluster {i} (Sin escalar)',
                legendgroup='no_scaling',
                showlegend=True
            ),
            row=1, col=2
        )
    
    # A침adir centroides sin escalar
    fig.add_trace(
        go.Scatter(
            x=centroides_no_scaling[:, 0],
            y=centroides_no_scaling[:, 1],
            mode='markers',
            marker=dict(
                color='black',
                size=12,
                symbol='x',
                line=dict(width=2)
            ),
            name='Centroides (Sin escalar)',
            legendgroup='no_scaling',
            showlegend=True
        ),
        row=1, col=2
    )
    
    # 3. K-means con StandardScaler (abajo a la izquierda)
    scaler_std = StandardScaler()
    X_std_scaled = scaler_std.fit_transform(X)
    kmeans_std = KMeans(n_clusters=n_clusters, random_state=random_state)
    clusters_std = kmeans_std.fit_predict(X_std_scaled)
    # Transformar los centroides de vuelta al espacio original
    centroides_std = scaler_std.inverse_transform(kmeans_std.cluster_centers_)
    
    for i in range(n_clusters):
        mask = clusters_std == i
        fig.add_trace(
            go.Scatter(
                x=X[mask, 0],  # Usar coordenadas originales para comparaci칩n
                y=X[mask, 1],
                mode='markers',
                marker=dict(color=colores[i % len(colores)], size=8),
                name=f'Cluster {i} (StandardScaler)',
                legendgroup='std_scaling',
                showlegend=True
            ),
            row=2, col=1
        )
    
    # A침adir centroides de StandardScaler
    fig.add_trace(
        go.Scatter(
            x=centroides_std[:, 0],
            y=centroides_std[:, 1],
            mode='markers',
            marker=dict(
                color='black',
                size=12,
                symbol='x',
                line=dict(width=2)
            ),
            name='Centroides (StandardScaler)',
            legendgroup='std_scaling',
            showlegend=True
        ),
        row=2, col=1
    )
    
    # 4. K-means con MinMaxScaler (abajo a la derecha)
    scaler_minmax = MinMaxScaler()
    X_minmax_scaled = scaler_minmax.fit_transform(X)
    kmeans_minmax = KMeans(n_clusters=n_clusters, random_state=random_state)
    clusters_minmax = kmeans_minmax.fit_predict(X_minmax_scaled)
    # Transformar los centroides de vuelta al espacio original
    centroides_minmax = scaler_minmax.inverse_transform(kmeans_minmax.cluster_centers_)
    
    for i in range(n_clusters):
        mask = clusters_minmax == i
        fig.add_trace(
            go.Scatter(
                x=X[mask, 0],  # Usar coordenadas originales para comparaci칩n
                y=X[mask, 1],
                mode='markers',
                marker=dict(color=colores[i % len(colores)], size=8),
                name=f'Cluster {i} (MinMaxScaler)',
                legendgroup='minmax_scaling',
                showlegend=True
            ),
            row=2, col=2
        )
    
    # A침adir centroides de MinMaxScaler
    fig.add_trace(
        go.Scatter(
            x=centroides_minmax[:, 0],
            y=centroides_minmax[:, 1],
            mode='markers',
            marker=dict(
                color='black',
                size=12,
                symbol='x',
                line=dict(width=2)
            ),
            name='Centroides (MinMaxScaler)',
            legendgroup='minmax_scaling',
            showlegend=True
        ),
        row=2, col=2
    )
    
    # Actualizar dise침o y etiquetas
    fig.update_layout(
        title=f'Comparaci칩n de M칠todos de Escalado para Clustering K-means ({columna_x} vs {columna_y})',
        height=800,
        width=1000,
        legend=dict(
            groupclick="toggleitem",
            tracegroupgap=5
        )
    )
    
    # Actualizar ejes con mismo t칤tulo en todos los subplots
    fig.update_xaxes(title_text=columna_x, row=1, col=1)
    fig.update_xaxes(title_text=columna_x, row=1, col=2)
    fig.update_xaxes(title_text=columna_x, row=2, col=1)
    fig.update_xaxes(title_text=columna_x, row=2, col=2)
    
    fig.update_yaxes(title_text=columna_y, row=1, col=1)
    fig.update_yaxes(title_text=columna_y, row=1, col=2)
    fig.update_yaxes(title_text=columna_y, row=2, col=1)
    fig.update_yaxes(title_text=columna_y, row=2, col=2)
    
    return fig
````

````python
def plot_dendrogram(model, **kwargs):
    """
    Visualiza el dendrograma de un modelo de clustering jer치rquico (AgglomerativeClustering).
    
    Esta funci칩n crea una matriz de enlace (linkage matrix) a partir de un modelo
    de clustering jer치rquico ya entrenado y visualiza su dendrograma, mostrando
    c칩mo se fueron agrupando las observaciones.

    Fuente: https://scikit-learn.org/stable/auto_examples/cluster/plot_agglomerative_dendrogram.html
    
    Args:
        model (AgglomerativeClustering): Modelo de clustering jer치rquico ya entrenado
            con compute_distances=True o con el m칠todo .fit() aplicado a los datos.
        **kwargs: Argumentos adicionales para pasar a la funci칩n dendrogram de scipy.
            Ejemplos: truncate_mode, p, leaf_font_size, color_threshold, etc.
    
    Returns:
        None: La funci칩n muestra el dendrograma pero no retorna ning칰n valor.
        
    Nota:
        El modelo debe tener los atributos children_ y distances_ disponibles
        (aseg칰rate de haber usado compute_distances=True al crear el modelo).
    """
    # Inicializamos un array para contar muestras bajo cada nodo
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    
    # Calculamos el n칰mero de muestras en cada fusi칩n de clusters
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # Es un nodo hoja (muestra original)
            else:
                current_count += counts[child_idx - n_samples]  # Es un subcluster
        counts[i] = current_count
    
    # Creamos la matriz de enlace con la estructura requerida por scipy
    linkage_matrix = np.column_stack(
        [model.children_, model.distances_, counts]
    ).astype(float)
    
    # Visualizamos el dendrograma con la matriz de enlace creada
    dendrogram(linkage_matrix, **kwargs)
````

````python
def visualizar_clusters_plotly(
    df: pd.DataFrame, 
    columna_x: str, 
    columna_y: str, 
    columna_cluster: str, 
    etiqueta_x: str = None, 
    etiqueta_y: str = None,
    titulo: str = 'Clusters resultantes',
    mostrar: bool = True
) -> go.Figure:
    """
    Visualiza clusters usando Plotly con un gr치fico de dispersi칩n interactivo.
    
    Args:
        df (pd.DataFrame): DataFrame con los datos
        columna_x (str): Nombre de la columna para el eje X
        columna_y (str): Nombre de la columna para el eje Y
        columna_cluster (str): Nombre de la columna que contiene los clusters
        etiqueta_x (str, optional): Etiqueta personalizada para el eje X
        etiqueta_y (str, optional): Etiqueta personalizada para el eje Y
        titulo (str, optional): T칤tulo del gr치fico
        mostrar (bool, optional): Si es True, muestra el gr치fico
        
    Returns:
        go.Figure: Figura con el gr치fico de dispersi칩n
    """
    # Crear una copia para no modificar el dataframe original
    df_plot = df.copy()
    
    # Asegurar que la columna de cluster es string para mejor visualizaci칩n
    df_plot[columna_cluster] = df_plot[columna_cluster].astype(str)
    
    # Usar las etiquetas proporcionadas o los nombres de columnas por defecto
    labels = {
        'x': etiqueta_x if etiqueta_x else columna_x,
        'y': etiqueta_y if etiqueta_y else columna_y
    }
    
    # Crear el gr치fico con Plotly
    fig = px.scatter(
        df_plot, 
        x=columna_x, 
        y=columna_y, 
        color=columna_cluster,
        color_discrete_sequence=px.colors.qualitative.Set1,
        opacity=0.7,
        labels=labels,
        title=titulo
    )
    
    # Actualizar dise침o del gr치fico
    fig.update_layout(
        width=800, 
        height=500,
        coloraxis_colorbar=dict(
            title='Cluster'
        )
    )
        
    return fig
````

````python
def visualizar_distribucion_clusters(
    df: pd.DataFrame, 
    columna_cluster: str, 
    columna_clase: str, 
    list_nombres_clases: list = None, 
    colores: list = None, 
    titulo: str = 'Distribuci칩n de clases reales en cada cluster',
) -> go.Figure:
    """
    Crea una visualizaci칩n de la distribuci칩n de clases en cada cluster con Plotly.
    
    Args:
        df (pd.DataFrame): DataFrame con los datos
        columna_cluster (str): Nombre de la columna que contiene los clusters
        columna_clase (str): Nombre de la columna que contiene las clases reales
        list_nombres_clases (list, optional): Lista con los nombres de las clases
        colores (list, optional): Lista de colores en formato hexadecimal
        titulo (str, optional): T칤tulo del gr치fico
        
    Returns:
        go.Figure: Figura con los gr치ficos de distribuci칩n
    """
    
    # Si no se proporcionan nombres de clases, usamos los valores 칰nicos de la columna clase
    if list_nombres_clases is None:
        list_nombres_clases = df[columna_clase].unique()
        
    # Si no se proporcionan colores, usamos la paleta Set1
    if colores is None:
        colores = ['#e41a1c', '#377eb8', '#4daf4a', '#984ea3', '#ff7f00']
    
    # Obtener la distribuci칩n de clases en cada cluster (valores absolutos)
    cluster_class_counts_abs = pd.crosstab(
        df[columna_cluster], 
        df[columna_clase]
    )
    
    # Calcular la distribuci칩n porcentual de clases en cada cluster
    cluster_class_counts = cluster_class_counts_abs.div(cluster_class_counts_abs.sum(axis=1), axis=0) * 100

    # Crear figura con subplots
    fig = make_subplots(
        rows=1, cols=2, 
        subplot_titles=('Distribuci칩n absoluta', 'Distribuci칩n porcentual (%)'),
        specs=[[{'type': 'bar'}, {'type': 'bar'}]]
    )

    # A침adir barras para cada clase en valores absolutos
    for i, clase in enumerate(list_nombres_clases):
        if clase in cluster_class_counts_abs.columns:
            fig.add_trace(
                go.Bar(
                    x=[f'Cluster {cluster}' for cluster in cluster_class_counts_abs.index],
                    y=cluster_class_counts_abs[clase],
                    name=clase,
                    marker_color=colores[i % len(colores)],
                    text=cluster_class_counts_abs[clase],
                    textposition='auto',
                ),
                row=1, col=1
            )

    # A침adir barras para cada clase en porcentajes
    for i, clase in enumerate(list_nombres_clases):
        if clase in cluster_class_counts.columns:
            fig.add_trace(
                go.Bar(
                    x=[f'Cluster {cluster}' for cluster in cluster_class_counts.index],
                    y=cluster_class_counts[clase],
                    name=clase,
                    marker_color=colores[i % len(colores)],
                    text=[f'{val:.1f}%' for val in cluster_class_counts[clase]],
                    textposition='auto',
                    showlegend=False
                ),
                row=1, col=2
            )

    # Actualizar layout
    fig.update_layout(
        title=titulo,
        barmode='group',
        height=500,
        width=1000,
        legend=dict(
            x=0.5,
            y=1.15,
            xanchor='center',
            orientation='h'
        )
    )

    fig.update_yaxes(title_text='N칰mero de observaciones', row=1, col=1)
    fig.update_yaxes(title_text='Porcentaje (%)', row=1, col=2)
        
    return fig
````

````python
def visualizar_pca_varianza(
    datos: pd.DataFrame, 
    aplicar_escalado: bool = True,
    umbral_varianza: float = 0.95,
    random_state: int = 42,
    titulo: str = "Scree plot y varianza acumulada (PCA)"
) -> go.Figure:
    """
    Visualiza la varianza explicada por cada componente principal y la varianza acumulada.
    
    Args:
        datos (pd.DataFrame): DataFrame con los datos a los que aplicar PCA
        aplicar_escalado (bool): Si True, aplica StandardScaler antes de PCA
        umbral_varianza (float): Umbral para marcar en el gr치fico (ej: 0.95 para 95%)
        random_state (int): Semilla para reproducibilidad
        titulo (str): T칤tulo del gr치fico
        
    Returns:
        go.Figure: Figura de Plotly con el gr치fico de varianza de PCA
    """
    # Preparamos los datos
    if aplicar_escalado:
        scaler = StandardScaler()
        X_scaled = scaler.fit_transform(datos)
    else:
        X_scaled = datos.values
    
    # Aplicamos PCA
    pca = PCA(random_state=random_state)
    pca.fit(X_scaled)
    
    # Obtenemos los datos de varianza
    varianza = pca.explained_variance_ratio_
    varianza_acum = np.cumsum(varianza)
    componentes = list(range(1, len(varianza)+1))
    
    # Creamos la figura con dos ejes Y
    fig = make_subplots(specs=[[{"secondary_y": True}]])
    
    # A침adimos el gr치fico de barras para la varianza individual
    fig.add_trace(
        go.Bar(
            x=componentes,
            y=varianza,
            name="Varianza por componente",
            marker_color='rgb(55, 83, 109)',
            opacity=0.7,
            hovertemplate='Componente %{x}<br>Varianza explicada: %{y:.3f}<extra></extra>'
        ),
        secondary_y=False
    )
    
    # A침adimos la l칤nea para la varianza acumulada
    fig.add_trace(
        go.Scatter(
            x=componentes,
            y=varianza_acum,
            name="Varianza acumulada",
            mode='lines+markers',
            marker=dict(color='red'),
            line=dict(width=3),
            hovertemplate='Componente %{x}<br>Varianza acumulada: %{y:.3f}<extra></extra>'
        ),
        secondary_y=True
    )
    
    # Configuramos los ejes y el dise침o
    fig.update_layout(
        title=titulo,
        xaxis=dict(
            title="N칰mero de componentes principales",
            tickmode='linear',
            tick0=1,
            dtick=1
        ),
        legend=dict(
            orientation="h",
            yanchor="bottom",
            y=1.02,
            xanchor="center",
            x=0.5
        ),
        hovermode="x unified",
        width=900,
        height=600,
        template="plotly_white",
    )
    
    # Configuramos los t칤tulos de los ejes Y
    fig.update_yaxes(title_text="Proporci칩n de varianza explicada", secondary_y=False)
    fig.update_yaxes(
        title_text="Varianza acumulada", 
        secondary_y=True,
        range=[0, 1.05],
        tickformat=".0%"
    )
    
    # A침adimos l칤nea de referencia en el umbral de varianza acumulada
    component_threshold = next((i for i, v in enumerate(varianza_acum) if v >= umbral_varianza), 
                              len(varianza_acum))
    
    if component_threshold < len(varianza_acum):
        fig.add_vline(
            x=component_threshold + 1,  # +1 porque los componentes empiezan en 1
            line_dash="dash",
            line_color="green",
            annotation_text=f"{umbral_varianza*100:.0f}% de varianza con {component_threshold + 1} componentes",
            annotation_position="top right"
        )
    
    return fig
````

````python
def visualizar_proyeccion_2d(
    df: pd.DataFrame, 
    columna_x: str = 'Componente 1', 
    columna_y: str = 'Componente 2', 
    columna_color: str = None,
    etiqueta_x: str = None, 
    etiqueta_y: str = None,
    titulo: str = 'Proyecci칩n 2D',
    metodo: str = 'PCA',
    paleta_colores: list = None,
    ancho: int = 800,
    alto: int = 600,
    tamano_punto: int = 8,
    mostrar_leyenda: bool = True,
    opacidad: float = 0.8
) -> go.Figure:
    """
    Visualiza proyecciones 2D de t칠cnicas de reducci칩n de dimensionalidad como PCA, UMAP o t-SNE.
    
    Args:
        df (pd.DataFrame): DataFrame con los datos proyectados (debe contener al menos las columnas columna_x y columna_y)
        columna_x (str): Nombre de la columna para el eje X (por defecto: 'Componente 1')
        columna_y (str): Nombre de la columna para el eje Y (por defecto: 'Componente 2')
        columna_color (str): Nombre de la columna para colorear los puntos (opcional)
        etiqueta_x (str): Etiqueta personalizada para el eje X
        etiqueta_y (str): Etiqueta personalizada para el eje Y
        titulo (str): T칤tulo del gr치fico
        metodo (str): M칠todo de reducci칩n de dimensionalidad ('PCA', 'UMAP', 't-SNE', etc.)
        paleta_colores (list): Lista personalizada de colores (si None, usa la paleta Set1)
        ancho (int): Ancho del gr치fico en p칤xeles
        alto (int): Altura del gr치fico en p칤xeles
        tamano_punto (int): Tama침o de los puntos en el gr치fico
        mostrar_leyenda (bool): Si es True, muestra la leyenda
        opacidad (float): Opacidad de los puntos (0-1)
        
    Returns:
        go.Figure: Figura de Plotly con la proyecci칩n 2D
    """
    # Verificar que las columnas existen en el dataframe
    for col in [columna_x, columna_y]:
        if col not in df.columns:
            raise ValueError(f"La columna '{col}' no existe en el DataFrame proporcionado")
    
    if columna_color and columna_color not in df.columns:
        raise ValueError(f"La columna de color '{columna_color}' no existe en el DataFrame")
    
    # Crear una copia para no modificar el dataframe original
    df_plot = df.copy()
    
    # Asegurar que la columna de cluster sea del tipo adecuado para plotly
    if columna_color and 'cluster' in columna_color.lower():
        # Convertir columna a tipo categ칩rico para evitar problemas con valores num칠ricos
        df_plot[columna_color] = df_plot[columna_color].astype(str)
    
    # Determinar las etiquetas de los ejes
    if not etiqueta_x:
        if 'PC' in columna_x:
            etiqueta_x = f'Componente {columna_x[2:]}'
        else:
            etiqueta_x = columna_x
            
    if not etiqueta_y:
        if 'PC' in columna_y:
            etiqueta_y = f'Componente {columna_y[2:]}'
        else:
            etiqueta_y = columna_y
    
    # Configurar el t칤tulo completo
    titulo_completo = f'{titulo} ({metodo})'
    
    # Configurar las opciones para el scatter plot
    scatter_kwargs = {
        'x': columna_x,
        'y': columna_y,
        'opacity': opacidad,
        'title': titulo_completo,
        'labels': {columna_x: etiqueta_x, columna_y: etiqueta_y}
    }
    
    # Si hay una columna para colorear los puntos
    if columna_color:
        scatter_kwargs['color'] = columna_color
        
        # Verificar si es num칠rica para usar color continuo o discreto
        if pd.api.types.is_numeric_dtype(df_plot[columna_color]) and 'cluster' not in columna_color.lower():
            # Si es num칠rica y no es un cluster, usar escala continua
            scatter_kwargs['color_continuous_scale'] = 'Viridis'
        else:
            # Si es categ칩rica o un cluster, usar colores discretos
            # Usar la paleta personalizada o la predeterminada
            if paleta_colores:
                scatter_kwargs['color_discrete_sequence'] = paleta_colores
            else:
                scatter_kwargs['color_discrete_sequence'] = px.colors.qualitative.Set1
            
            # No usar 'identity' que puede causar problemas con valores num칠ricos
            scatter_kwargs.pop('color_discrete_map', None)
    else:
        # Si no hay una columna de color, usar un 칰nico color
        scatter_kwargs['color_discrete_sequence'] = ['#636EFA']
    
    # Crear el gr치fico con Plotly Express
    fig = px.scatter(df_plot, **scatter_kwargs)
    
    # Actualizar el dise침o del gr치fico
    layout_updates = {
        'width': ancho,
        'height': alto,
        'template': 'plotly_white',
        'xaxis_title': etiqueta_x,
        'yaxis_title': etiqueta_y,
        'showlegend': mostrar_leyenda
    }
    
    # Si hay una columna de color, ajustar el t칤tulo de la leyenda
    if columna_color:
        layout_updates['legend_title_text'] = columna_color.capitalize()
    
    # Actualizar el tama침o de los marcadores
    fig.update_traces(marker=dict(size=tamano_punto))
    
    # Aplicar las actualizaciones de dise침o
    fig.update_layout(**layout_updates)
    
    return fig

def visualizar_proyeccion_3d(
    df: pd.DataFrame, 
    columna_x: str = 'Componente 1', 
    columna_y: str = 'Componente 2', 
    columna_z: str = 'Componente 3',
    columna_color: str = None,
    etiqueta_x: str = None, 
    etiqueta_y: str = None,
    etiqueta_z: str = None,
    titulo: str = 'Proyecci칩n 3D',
    metodo: str = 'PCA',
    paleta_colores: list = None,
    ancho: int = 900,
    alto: int = 700,
    tamano_punto: int = 6,
    mostrar_leyenda: bool = True,
    opacidad: float = 0.8,
    angulo_camara: dict = None
) -> go.Figure:
    """
    Visualiza proyecciones 3D de t칠cnicas de reducci칩n de dimensionalidad como PCA, UMAP o t-SNE.
    
    Args:
        df (pd.DataFrame): DataFrame con los datos proyectados 
        columna_x (str): Nombre de la columna para el eje X (por defecto: 'Componente 1')
        columna_y (str): Nombre de la columna para el eje Y (por defecto: 'Componente 2')
        columna_z (str): Nombre de la columna para el eje Z (por defecto: 'Componente 3')
        columna_color (str): Nombre de la columna para colorear los puntos (opcional)
        etiqueta_x (str): Etiqueta personalizada para el eje X
        etiqueta_y (str): Etiqueta personalizada para el eje Y
        etiqueta_z (str): Etiqueta personalizada para el eje Z
        titulo (str): T칤tulo del gr치fico
        metodo (str): M칠todo de reducci칩n de dimensionalidad ('PCA', 'UMAP', 't-SNE', etc.)
        paleta_colores (list): Lista personalizada de colores (si None, usa la paleta Set1)
        ancho (int): Ancho del gr치fico en p칤xeles
        alto (int): Altura del gr치fico en p칤xeles
        tamano_punto (int): Tama침o de los puntos en el gr치fico
        mostrar_leyenda (bool): Si es True, muestra la leyenda
        opacidad (float): Opacidad de los puntos (0-1)
        angulo_camara (dict): Diccionario con los 치ngulos de la c치mara (eye, center, up)
        
    Returns:
        go.Figure: Figura de Plotly con la proyecci칩n 3D
    """
    # Verificar que las columnas existen en el dataframe
    for col in [columna_x, columna_y, columna_z]:
        if col not in df.columns:
            raise ValueError(f"La columna '{col}' no existe en el DataFrame proporcionado")
    
    if columna_color and columna_color not in df.columns:
        raise ValueError(f"La columna de color '{columna_color}' no existe en el DataFrame")
    
    # Crear una copia para no modificar el dataframe original
    df_plot = df.copy()
    
    # Determinar las etiquetas de los ejes
    def generar_etiqueta(columna, etiqueta_predefinida):
        if etiqueta_predefinida:
            return etiqueta_predefinida
        elif 'PC' in columna:
            return f'Componente {columna[2:]}'
        else:
            return columna
    
    etiqueta_x = generar_etiqueta(columna_x, etiqueta_x)
    etiqueta_y = generar_etiqueta(columna_y, etiqueta_y)
    etiqueta_z = generar_etiqueta(columna_z, etiqueta_z)
    
    # Configurar el t칤tulo completo
    titulo_completo = f'{titulo} ({metodo})'
    
    # Crear la figura 3D
    fig = go.Figure()
    
    # Determinar los colores para los puntos
    if columna_color:
        # Si es num칠rica, usamos escala continua
        if pd.api.types.is_numeric_dtype(df_plot[columna_color]):
            fig.add_trace(
                go.Scatter3d(
                    x=df_plot[columna_x],
                    y=df_plot[columna_y],
                    z=df_plot[columna_z],
                    mode='markers',
                    marker=dict(
                        size=tamano_punto,
                        color=df_plot[columna_color],
                        opacity=opacidad,
                        colorscale='Viridis',
                        showscale=True,
                        colorbar=dict(title=columna_color)
                    ),
                    hovertemplate=(
                        f"{etiqueta_x}: %{{x:.2f}}<br>"
                        f"{etiqueta_y}: %{{y:.2f}}<br>"
                        f"{etiqueta_z}: %{{z:.2f}}<br>"
                        f"{columna_color}: %{{marker.color}}<extra></extra>"
                    )
                )
            )
        else:
            # Para variables categ칩ricas, usamos colores discretos
            categorias = df_plot[columna_color].unique()
            
            # Seleccionar paleta de colores
            if paleta_colores:
                colores = paleta_colores
            else:
                colores = px.colors.qualitative.Set1
            
            # Crear un trace para cada categor칤a
            for i, categoria in enumerate(categorias):
                mask = df_plot[columna_color] == categoria
                fig.add_trace(
                    go.Scatter3d(
                        x=df_plot.loc[mask, columna_x],
                        y=df_plot.loc[mask, columna_y],
                        z=df_plot.loc[mask, columna_z],
                        mode='markers',
                        marker=dict(
                            size=tamano_punto,
                            color=colores[i % len(colores)],
                            opacity=opacidad
                        ),
                        name=str(categoria),
                        hovertemplate=(
                            f"{etiqueta_x}: %{{x:.2f}}<br>"
                            f"{etiqueta_y}: %{{y:.2f}}<br>"
                            f"{etiqueta_z}: %{{z:.2f}}<br>"
                            f"{columna_color}: {categoria}<extra></extra>"
                        )
                    )
                )
    else:
        # Si no hay columna de color, usar un 칰nico color
        fig.add_trace(
            go.Scatter3d(
                x=df_plot[columna_x],
                y=df_plot[columna_y],
                z=df_plot[columna_z],
                mode='markers',
                marker=dict(
                    size=tamano_punto,
                    color='#636EFA',
                    opacity=opacidad
                ),
                hovertemplate=(
                    f"{etiqueta_x}: %{{x:.2f}}<br>"
                    f"{etiqueta_y}: %{{y:.2f}}<br>"
                    f"{etiqueta_z}: %{{z:.2f}}<extra></extra>"
                )
            )
        )
    
    # Actualizar el dise침o del gr치fico
    layout_updates = {
        'width': ancho,
        'height': alto,
        'title': titulo_completo,
        'scene': dict(
            xaxis_title=etiqueta_x,
            yaxis_title=etiqueta_y,
            zaxis_title=etiqueta_z,
            # Configurar aspecto para que sea c칰bico
            aspectmode='cube'
        ),
        'template': 'plotly_white',
        'showlegend': mostrar_leyenda
    }
    
    # Configurar 치ngulo de c치mara si se proporciona
    if angulo_camara:
        layout_updates['scene']['camera'] = angulo_camara
    
    # Aplicar las actualizaciones de dise침o
    fig.update_layout(**layout_updates)
    
    return fig
````

### Creaci칩n del modelo
+ En este caso utilizaremos las 4 variables que se utilizan para medir los p칠talos y s칠palos de cada flor con el objetivo de encontrar los diferentes grupos que se puedan formar.
+ Dentro del paradigma de muchos modelos de clustering es responder a la pregunta "쮺u치ntos clusters se podr치n encontrar en los datos?", ya que hay que definir este n칰mero previamente.
+ K-Means no te dice cu치l es el n칰mero de clusters m치s 칩ptimo.
+ No utilizaremos el m칠todo de `train_test_split` porque normalmente los modelos de clustering no se suelen evaluar.

from sklearn.cluster import KMeans

Los hiperpar치metros m치s recomendados son:
+ `n_clusters`: representa el valor `k`, que es el n칰mero de clusters.
+ `init`: m칠todo de inicializaci칩n de los centroides, el valor m치s com칰n (y por defecto) es `k-means++`. Tambi칠n puede ser `random`.
+ `max_iter`: n칰mero de iteraciones m치ximas que har치 hasta encontrar los centroides m치s 칩ptimos.

#### Elbow Method
La manera tradicional de encontrar el n칰mero 칩ptimo de clusters es el famoso Elbow Method (M칠todo del Codo), a trav칠s de un gr치fico viendo c칩mo decrede un valor llamado `inertia`.

$$\text{Inertia} = \sum_{i=1}^{n} \min_{\mu_j \in C} |x_i - \mu_j|^2$$

Donde:
+ $x_i$ representa cada punto de datos
+ $\mu_j$ representa cada centroide
+ $C$ es el conjunto de todos los centroides
+ $|x_i - \mu_j|^2$ es el cuadrado de la distancia euclidiana entre el punto $x_i$ y el centroide $\mu_j$
+ $\min_{\mu_j \in C}$ indica que para cada punto tomamos la distancia m칤nima a cualquiera de los centroides

### Importancia del escalado o normalizar los datos

El escalado de datos es crucial en clustering por estos motivos:
+ **Evita el sesgo por magnitudes**: Algoritmos como K-means son sensibles a variables con diferentes escalas, dando m치s peso a las que tienen valores m치s grandes
+ **Mejora c치lculos de distancia**: Las m칠tricas de distancia funcionan correctamente cuando todas las variables tienen escalas comparables
+ **Acelera la convergencia**: Los algoritmos convergen m치s r치pido y de forma m치s estable con datos escalados

Principales t칠cnicas:
+ `StandardScaler` (escalado):
  + Transforma variables para tener media 0 y desviaci칩n est치ndar 1
  + Ideal para: Distribuciones aproximadamente normales o datos con outliers
  + F칩rmula: `z = (x - 풮)/픢`
$$z = \frac{x - \mu}{\sigma}$$

+ `MinMaxScaler` (normalizar):
  + Escala datos al rango [0,1]
  + Ideal para: Preservar relaciones exactas entre valores o distribuciones no gaussianas
  + F칩rmula: `x_scaled = (x - x_min)/(x_max - x_min)`

$$x_{scaled} = \frac{x - x_{min}}{x_{max} - x_{min}}$$
