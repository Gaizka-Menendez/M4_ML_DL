```python
# Instalación
# !pip install pycaret[full]

import pandas as pd
from pycaret.datasets import get_data
```

## Modelo de Clasificación

```python
from pycaret import classification as cl
```

```python
# Cargar dataset de ejemplo para clasificación
data = get_data('juice')
print(data.head())
```

```python
# Setup del experimento
exp_clf = cl.setup(data=data, target='Purchase', session_id=123, html=True)
```

```python
# Comparar modelos
best_model = cl.compare_models()
```

```python
best_model
```

```python
# Crear un modelo específico (Random Forest)
rf = cl.create_model('rf')
```

```python
# Evaluar modelo
cl.evaluate_model(rf)
```

```python
# Optimizar hiperparámetros
tuned_rf = cl.tune_model(rf)
```

```python
# Interpretar modelo
cl.plot_model(tuned_rf, plot='feature')
```

```python
# Validación cruzada y resultados
cv_results = cl.pull()
cv_results
```

```python
# Guardar modelo
cl.save_model(tuned_rf, 'rf_model_juice')
```

```python
# Cargar modelo guardado
loaded_rf = cl.load_model('rf_model_juice')
```

## Modelo de Regresión

```python
# Regresión con PyCaret
from pycaret import regression as rg
```

```python
data_reg = get_data('insurance')
data_reg.head()
```

```python
exp_reg = rg.setup(data=data_reg, target='charges', session_id=123, html=True)
```

```python
best_reg = rg.compare_models()
```

```python
best_model_rg = rg.create_model('rf')
```

```python
tuned_lr = rg.tune_model(best_model_rg)
```

```python
rg.evaluate_model(tuned_lr)
```

## Otras funcionalidades

```python
# Crear una app para hacer la inferencia con el modelo
rg.create_app(tuned_lr)
```

```python
# Creación de un dashboard de interpretabilidad
rg.dashboard(tuned_lr)
```
